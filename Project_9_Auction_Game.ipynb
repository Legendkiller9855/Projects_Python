```python
#AUCTION
bidders={}
def function():
  name_of_bidder = input("Enter the name of the person who is bidding?")
  amount_to_be_bid = int(input(f"Enter the amount to be bid? : $"))
  bidders[name_of_bidder]=amount_to_be_bid
  more_bidders = input("Are there any more bidders? Type 'yes' or 'no'.")
  if more_bidders == "yes":
    print("\n"*100)
    function()

function()

def highest_bidder():
  highest_bid = 0
  winner = ""
  for bidder in bidders:
    bid_amount = bidders[bidder]
    if bid_amount > highest_bid:
      highest_bid = bid_amount
      winner = bidder
  print(f"The winner is {winner} with a bid of ${highest_bid}")

highest_bidder()
```
